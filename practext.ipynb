```python
text1=open("/kaggle/input/textanalysis/A7Input1.txt",'r')
text1=text1.read()
```

```python
text1=text1.lower()
```

```python
text2=open('/kaggle/input/textanalysis/A7Input2.txt','r')
text2=text2.read()
text2=text2.lower()
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
import string
x=string.punctuation +"{}[]()"
for i in x:
    text2=text2.replace(i," ")
    text1=text1.replace(i," ")
```

```python
ps=PorterStemmer()
```

```python
tok1=word_tokenize(text1)
```

```python
tok2=word_tokenize(text2)
```

```python
for i in tok2:
    print(ps.stem(i))
```

```python
for i in tok1:
    print(ps.stem(i))
```

```python
w=WordNetLemmatizer()
for i in tok1:
    print(w.lemmatize(i))
```

```python
x1=sent_tokenize(text1)
```

```python
x2=sent_tokenize(text2)
```

```python
x2
```

```python
l=[text1,text2]
```

```python
l1=[x1[0],x2[0]]
```

```python
l1
```

```python
l={}
for i in tok1:
    if i not in l:
        l[i]=1
    else:
        l[i]+=1
for i,j in l.items():
    l[i]=l[i]/len(tok1)
```

```python
l
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
l=TfidfVectorizer()
```

```python
res=l.fit_transform(l1)
```

```python
arr=res.toarray()
```

```python
arr
```

```python
feature_name=l.get_feature_names_out()
```

```python
for i,j in enumerate(l1):
    for z,k in enumerate(feature_name):
        score=arr[i][z]
        print(k," ",score)
```
