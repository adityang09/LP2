```python
from nltk.tokenize import sent_tokenize,word_tokenize
```

```python
tok1=word_tokenize(t1)
```

```python
l={}
for i in tok1:
    if i not in l:
        l[i]=1
    else:
        l[i]+=1
l        
for i,j in l.items():
    l[i]=l[i]/len(tok1)
```

```python
l
```

```python
sentence_tokenized = sent_tokenize(t2)
print(sentence_tokenized)
```

```python
sentence_tokenized[0]
```

```python
sentene_tokenize= sent_tokenize(t1)
print(sentence_tokenized)
```

```python
L=[sentene_tokenize[0],sentence_tokenized[0]]
```

```python
L
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer(stop_words='english')
```

```python
result = tf.fit_transform(L)
print(result)
```

```python
feature_names = tf.get_feature_names_out()
```

```python
tfidf_array=result.toarray()
```

```python
for doc_idx, doc in enumerate(L):
    print(f"Document {doc_idx+1}:")
    for word_idx, word in enumerate(feature_names):
        tfidf_score = tfidf_array[doc_idx][word_idx]
        print(f"  {word}: {tfidf_score:.4f}")
```
